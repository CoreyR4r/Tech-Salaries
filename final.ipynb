```python
import pandas as pd

def clean_csv(file_path, output_path):
    try:
        df = pd.read_csv(file_path)
        df_cleaned = df.replace("NA", pd.NA).dropna()
        
        df_cleaned.to_csv(output_path, index=False)
        print(f"Cleaned data saved to {output_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

input_file = r"C:\statistics excel\tech_companies_salary (2).csv"
output_file = r"C:\statistics excel\cleaned_output3.csv"

clean_csv(input_file, output_file)

```

```python
import pandas as pd

def sort_excel_by_title_and_compensation(input_file, output_file):

    try:
        df = pd.read_excel(input_file)
        if 'title' not in df.columns or 'totalyearlycompensation' not in df.columns:
            raise ValueError("The Excel file must contain 'title' and 'totalyearlycompensation' columns.")
        df_sorted = df.sort_values(by=['title', 'totalyearlycompensation'], ascending=[True, False])

        df_sorted.to_excel(output_file, index=False)
        print(f"Sorted data saved to {output_file}")
    except Exception as e:
        print(f"An error occurred: {e}")


input_excel = r"C:\statistics excel\Book1.xlsx" 
output_excel = r"C:\statistics excel\Book34.xlsx"  

sort_excel_by_title_and_compensation(input_excel, output_excel)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:\statistics excel\Races.xlsx"
data = pd.ExcelFile(file_path)
df = data.parse('Sheet1')

df_cleaned = df[['Asian', 'White', '2 or more', 'Black', 'Hispanic']]
df_melted = df_cleaned.melt(var_name="Race", value_name="Total Yearly Compensation")
df_melted.dropna(inplace=True)

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x="Race", y="Total Yearly Compensation", palette="Set2")

plt.title("Total Yearly Compensation by Race", fontsize=16)
plt.xlabel("Race", fontsize=14)
plt.ylabel("Total Yearly Compensation (USD)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:\statistics excel\Education.xlsx"
excel_data = pd.ExcelFile(file_path)

data = excel_data.parse('Sheet1')

cleaned_data = data[['Bachelor\'s Degree', 'Master\'s Degree', 'PHD', 'Some College', 'High School']]

reshaped_data = cleaned_data.melt(var_name='Education Level', value_name='Total Yearly Compensation')

reshaped_data = reshaped_data.dropna()

plt.figure(figsize=(10, 6))
plt.boxplot(
    [reshaped_data[reshaped_data['Education Level'] == level]['Total Yearly Compensation'] 
     for level in reshaped_data['Education Level'].unique()],
    labels=reshaped_data['Education Level'].unique(),
    patch_artist=True
)
plt.title('Box and Whisker Plot of Total Yearly Compensation by Education Level')
plt.ylabel('Total Yearly Compensation')
plt.xlabel('Education Level')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:\statistics excel\Gender.xlsx"
data = pd.read_excel(file_path, sheet_name='Sheet1')

data_melted = data.melt(value_vars=['Female', 'Male', 'Other'], var_name='Gender', value_name='Total Yearly Compensation')

data_melted = data_melted.dropna()

plt.figure(figsize=(10, 6))
sns.boxplot(data=data_melted, x='Gender', y='Total Yearly Compensation', palette='Set2')


plt.title('Box-and-Whisker Plots of Total Yearly Compensation by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Total Yearly Compensation ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```
